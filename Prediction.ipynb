#Exploratory Data Analysis (EDA) and Data Visualization

###Statistik Deskriptif

```python
df.describe()
```

###Univariate

```python
#distribusi pelanggan berdasarkan jenis kelamin.
gender_counts = df['gender'].value_counts()
plt.figure(figsize=(16,8))
gender_counts.plot(kind='bar', color=['blue', 'red'])
plt.title('Distribution of Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Rotasi label sumbu x jika perlu
plt.show()
```

```python
num_male = gender_counts['Male']
num_female = gender_counts['Female']

print(f"Jumlah laki-laki: {num_male}")
print(f"Jumlah perempuan: {num_female}")
```

```python
#distribusi pelanggan yang memiliki tanggungan.
dependent_count = df['Dependents'].value_counts()
plt.figure(figsize=(16,8))
plt.pie(dependent_count, labels=dependent_count.index, autopct='%1.1f%%', colors=['skyblue', 'lightgreen'])
plt.title('Distribution of Customers by Dependents')
plt.show()
```

###Bivariate

```python
#hubungan antara lama pelanggan menggunakan layanan (tenure) dengan biaya bulanan (MonthlyCharges).
plt.figure(figsize=(16, 8))
sns.scatterplot(x='tenure', y='MonthlyCharges', data=df, hue='tenure', palette='viridis')
plt.title('Scatter Plot of Tenure vs Monthly Charges')
plt.xlabel('Tenure')
plt.ylabel('Monthly Charges')
plt.legend(title='Tenure')
plt.show()
```

```python
#menghitung lama pelanggan dengan total biaya
# Menghitung korelasi antara 'tenure' dan 'TotalCharges'
corr_matrix = df[['tenure', 'MonthlyCharges']].corr()

plt.figure(figsize=(16, 8))
sns.heatmap(corr_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Heatmap of Tenure vs Total Charges')
plt.show()
```

###Multivariate

```python
#jenis kontrak ('Contract') dan biaya bulanan ('MonthlyCharges') berkaitan dengan keputusan pelanggan untuk berpindah ('Churn').
contract_churn_counts = df.groupby(['Contract', 'Churn']).size().unstack()

# Membuat bar chart
plt.figure(figsize=(10, 6))
contract_churn_counts.plot(kind='bar', stacked=True, color=['red', 'green'])
plt.title('Churn Status by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Churn', labels=['No', 'Yes'])
plt.show()
```

```python
#metode pembayaran ('PaymentMethod') dan biaya bulanan ('MonthlyCharges') berkaitan dengan keputusan pelanggan untuk berpindah ('Churn').
plt.figure(figsize=(10, 6))
sns.boxplot(x='PaymentMethod', y='MonthlyCharges', hue='Churn', data=df)
plt.title('Monthly Charges by Payment Method and Churn Status')
plt.xlabel('Payment Method')
plt.ylabel('Monthly Charges')
plt.legend(title='Churn', labels=['No', 'Yes'])
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
